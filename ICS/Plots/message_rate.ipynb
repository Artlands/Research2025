```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from matplotlib.figure import figaspect

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b','#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#1f78b4', '#33a02c']
markers = ['o', 's', 'v', 'D', 'P', 'X', 'd', 'p', 'x', 'h', 'H', '8']
```

```python
threshold = '1MB'

x_labels  = ['1B', '2B', '4B', '8B', '16B', '32B', '64B', '128B', '256B', '512B', '1K', '2K', '4KB']
openshmem = [0.145, 0.566, 1.133, 2.267, 4.536, 9.072, 17.389, 34.328, 67.479, 132.902, 249.123, 491.229, 855.883]
xbrtime   = [0.968, 2.043, 4.273, 8.836, 8.836, 8.836, 68.394, 116.926, 154.129, 181.720, 199.624, 209.982, 215.567]

# Find the index of the threshold
threshold_index = x_labels.index(threshold)

x_labels_select = x_labels[0:threshold_index+1]
openshmem_select = openshmem[0:threshold_index+1]
xbrtime_select = xbrtime[0:threshold_index+1]

labels = ['OpenSHMEM', 'xBGAS Runtime']
```

```python
w,h = figaspect(1/2)
fig, ax = plt.subplots(figsize=(8, 4))

plt.rcParams['font.size'] = 10
lsize = 12
msize = 8

ax.plot(x_labels_select, openshmem_select, label=labels[0], color=colors[0], marker=markers[0], markersize=msize)
ax.plot(x_labels_select, xbrtime_select, label=labels[1], color=colors[1], marker=markers[1], markersize=msize)

ax.yaxis.grid(linestyle='--') 

# ax.set_ylim([0, 1.19])
# Set y ticks to be in log scale
# ax.set_yscale('log')

# Change the size of ticks and x/y axis labels
ax.tick_params(axis='both', which='major')
ax.set_ylabel("Messsage Rate (MB/s)", size=lsize)
ax.set_xlabel("Payload Size", size=lsize)
ax.legend(frameon=False)

plt.tight_layout()

# plt.savefig(f'./figures/message_rate.pdf', dpi=300, bbox_inches='tight')
```
